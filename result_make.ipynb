```python
import csv
import heapq

def read_csv(file_path):
    with open(file_path, 'r') as file:
        reader = csv.DictReader(file)
        records = [row for row in reader]
    return records

csv_file_path = "/aceph/louisyuzhao/buddy2/linyangxiao/AMP/AMPGPT/result/reg_infer.csv"
records = read_csv(csv_file_path)

# 提取具有最小预测值的 10 条记录
min_records = heapq.nsmallest(25, records, key=lambda x: float(x["Prediction"]))

# 提取具有最大预测值的 10 条记录
max_records = heapq.nlargest(100, records, key=lambda x: float(x["Prediction"]))

# 输出结果
print("最小预测值的 10 条记录：")
for record in min_records:
    print(record)

print("\n最大预测值的 10 条记录：")
for record in max_records:
    print(record)

print("\n最大预测值的 10 条记录：")
for record in max_records:
    print(record)
def save_records_to_txt(records, file_path):
    with open(file_path, "w") as file:
        for record in records:
            sequence = record["Sequence"]
            prediction = record["Prediction"]
            file.write(f"{sequence}\t{prediction}\n")

min_records_file_path = "min_predictions.txt"
save_records_to_txt(min_records, min_records_file_path)

max_records_file_path = "max_predictions.txt"
save_records_to_txt(max_records, max_records_file_path)
```

```python
import csv
import heapq

def read_csv(file_path):
    with open(file_path, 'r') as file:
        reader = csv.reader(file)
        records = [row for row in reader]
    return records

def add_column_names(records):
    column_names = ["Sequence", "Prediction"]
    named_records = [dict(zip(column_names, row)) for row in records if float(row[1]) > 0]
    return named_records

csv_file_path = "/aceph/louisyuzhao/buddy2/linyangxiao/AMP/AMPGPT/Data/AMP/AMP.csv"
records = read_csv(csv_file_path)
named_records = add_column_names(records)

# 提取具有最小预测值的 10 条记录
min_records = heapq.nsmallest(10, named_records, key=lambda x: float(x["Prediction"]))

# 提取具有最大预测值的 10 条记录
max_records = heapq.nlargest(10, named_records, key=lambda x: float(x["Prediction"]))

# 输出结果
print("最小预测值的 10 条记录：")
for record in min_records:
    print(record)

print("\n最大预测值的 10 条记录：")
for record in max_records:
    print(record)

# 保存记录到 TXT 文件
min_records_file_path = "min_true.txt"
save_records_to_txt(min_records, min_records_file_path)

max_records_file_path = "max_true.txt"
save_records_to_txt(max_records, max_records_file_path)
```
